```python
#need to remove newline, tabs and multiple spaces
```

```python
df['summary'] = df['summary'].str.replace("\n", " ", regex = "False")
df['summary'] = df['summary'].str.replace("\t", " ", regex = "False")
print(df['summary'].str.count(r'\t').sum())
print(df['summary'].str.count(r'\n').sum())
```

```python
#combining into one function

def cleaner(text):
    text = str(text)

    #replace newlines and tabs
    text = text.replace("\n", " ")
    text = text.replace("\t", " ")

    #replace multiple spaces
    text = re.sub(r"\s+", " ", text)

    #strip trailing and leading spaces
    text = text.strip()

    return text
```

```python
df['title'] = df['title'].apply(cleaner)
df['summary'] = df['summary'].apply(cleaner)

print(df['title'].str.count(r'\n').sum())
print(df['summary'].str.count(r'\n').sum())
print(df['summary'].str.count(r'\t').sum())
print(df['title'].str.count(r'\t').sum())
print(df['summary'].str.count(r' {2,}').sum())
print(df['title'].str.count(r' {2,}').sum())
```

```python
df.head()
```

```python
df['document'] = df['title'] + " " + df['summary']

df.head()
```

```python
df = df[df['document'].str.strip() != ""]

```

```python
print(df.shape[0])
```

```python
df = df.drop_duplicates(subset = ['document'])

print(df.shape[0])
```
